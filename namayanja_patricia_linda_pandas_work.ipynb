# Creating series and comparing them

```python
# using pandas to compare the two series
series1 = pd.Series([4, 65, 436, 3, 9])
series2 = pd.Series([7, 0, 3, 897, 9])

# Compare the elements of the two series
comparison = series1 == series2

# Display the comparison result
print(comparison)
```

# Carrying out operations on series created

```python
# Created sample Series
series1 = pd.Series([2, 4, 6, 8, 14])
series2 = pd.Series([1, 3, 5, 7, 9])


addition = series1 + series2
print("Addition:")
print(addition)


subtraction = series1 - series2
print("\nSubtraction:")
print(subtraction)

multiplication = series1 * series2
print("\nMultiplication:")
print(multiplication)



division = series1 / series2
print("\nDivision:")
print(division)
```
